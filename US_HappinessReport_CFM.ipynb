```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Name of the CSV file
us_happiness18 = "data_sets/2018_US_Happiness_Report.csv"
states_coordinates = "data_sets/States_Coordinates.csv"
```

```python
# The correct encoding must be used to read the CSV in pandas
ushappy_df = pd.read_csv(us_happiness18)
coordinates_df = pd.read_csv(states_coordinates)
```

```python
# Preview of the 2018 US Happy Report DataFrame
ush18_df = pd.DataFrame(ushappy_df[["Overall Rank (1 Happiest)","State","Total Score"]])
ush18_df = ush18_df.rename(columns={"Overall Rank (1 Happiest)": "Happiness Rank 2018","Total Score":"Happiness Score 2018"})
ush18_df = ush18_df[["State","Happiness Rank 2018","Happiness Score 2018"]]
ush18_df.head()
# ush18_df.count()
# ush18_df.dtypes
```

```python
# Preview of the Country Coordinates Data Frame
sc_df = pd.DataFrame(coordinates_df[["latitude","longitude","name"]])
sc_df = sc_df.rename(columns={"latitude": "Latitude","longitude":"Longitude","name":"State"})
sc_df = sc_df[["State","Latitude","Longitude"]]
sc_df.head()
# sc_df.count()
# sc_df.dtypes
```

```python
# Merge the DataFrame
us_happy_df = pd.merge(ush18_df, sc_df, on="State", how="inner")
us_happy_df = us_happy_df[["State","Latitude","Longitude","Happiness Rank 2018","Happiness Score 2018"]]
us_happy_df
#us_happy_df.count()
```

```python
#Save File
us_happy_df.to_csv("data_sets/us_happy.csv")
```
