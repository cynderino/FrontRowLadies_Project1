```python
import pandas as pd

csvpath = "data_sets/2018_US_Happiness_Report.csv"
csvpath2 = "data_sets/us_ave_annual_sunshine.csv"
csvpath3 = "data_sets/us_ave_annual_temp.csv"

ushappiness_df = pd.read_csv(csvpath)
ussunshine_df = pd.read_csv(csvpath2)
ustemps_df = pd.read_csv(csvpath3)
```

```python
happiness_df = ushappiness_df.filter(['State','Overall Rank (1 Happiest)','Total Score'], axis=1)
happiness_df = happiness_df.sort_values('State', ascending=True)
happiness_df = happiness_df.set_index('State')
```

```python
happiness_df.head(5)
```

```python
sunshine_df = ussunshine_df.filter(['State','% Sun','Total Hours','Clear Days'], axis=1)
sunshine_df = sunshine_df.set_index('State')
```

```python
sunshine_df.head(5)
```

```python
temps_df = ustemps_df.filter(['State','Avg °F','Avg °C','Rank'], axis=1)
temps_df = temps_df.set_index('State')
```

```python
temps_df.head(5)
```

```python
ushappysunshine_df = pd.merge(happiness_df, sunshine_df, on="State", how="outer")
```

```python
ushappysunshine_df.head()
```

```python
allusdata_df = pd.merge(ushappysunshine_df, temps_df, on="State", how="outer")
```

```python
allusdata_df.head()
```

```python
allusdata_df.to_csv("Data_Sets/all_us_data.csv")
```
