```python
import pandas as pd
import numpy as np

csvpath = "Data_Sets/merged_whp_dep_coord.csv"

df = pd.read_csv(csvpath)
```

```python
df.head()
```

```python
df=df[["Country", "Country Code", "Latitude", "Longitude", "Happiness Rank 2016"]]
df.head()
```

```python
df=df.sort_values('Happiness Rank 2016')
len(df)
```

```python
top10_df = df.nsmallest(10, "Happiness Rank 2016")
top10_df
```

```python
bottom10_df = df.nlargest(10,"Happiness Rank 2016")
bottom10_df
```

```python
middle10_df = df.loc[(df["Happiness Rank 2016"] >= 64) & 
                     (df["Happiness Rank 2016"] < 74)]
middle10_df
```

```python
us_df = df.loc[(df["Country"] == "United States")]
us_df
```

```python
top_middle_bottom_df = top10_df.append(middle10_df)
top_middle_bottom_df = top_middle_bottom_df.append(bottom10_df)
top_middle_bottom_df = top_middle_bottom_df.append(us_df)
top_middle_bottom_df
```

```python
top_middle_bottom_df.columns
```

```python
top_middle_bottom_df = top_middle_bottom_df.reset_index(drop=True)
top_middle_bottom_df
```

```python
top_middle_bottom_df.to_csv("Data_Sets/Cities_input.csv")
```
