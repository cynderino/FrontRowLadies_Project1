```python
import pandas as pd


#import csv files
whpcsvpath = "Data_Sets/world_happy.csv"

whp_countries_df = pd.read_csv(whpcsvpath)

whp_countries_df.head()
```

```python
depscvpath = "Data_Sets/depression_by_country_2yr.csv"
depression_countries_df = pd.read_csv(depscvpath)

depression_countries_df.head()
```

```python
#merge on country
countries_df = pd.merge(whp_countries_df, depression_countries_df, on="Country", how="inner")
countries_df.head()
```

```python
#checked data, no Nans.
#export to csv

countries_df.to_csv("Data_Sets/whp_depression_merge.csv")
```
